```python
import numpy as np
import matplotlib.pyplot as plt
from utils import base_plot, add_plot_trajectory, EllipsoidTubeData2D
```

```python
# load numpy data
data_zoro_acados = np.load("solve_data_zoro_acados.npy", allow_pickle=True).item()
data_zoro_acados_custom_update = np.load(
    "solve_data_zoro_acados_custom_update.npy", allow_pickle=True
).item()
data_zero_order_gpmpc = np.load(
    "solve_data_zero_order_gpmpc.npy", allow_pickle=True
).item()
```

```python
plt.plot(
    data_zoro_acados["X"][:, 0],
    data_zoro_acados["X"][:, 1],
    label="zoro_acados",
    linewidth=5,
)
plt.plot(
    data_zoro_acados_custom_update["X"][:, 0],
    data_zoro_acados_custom_update["X"][:, 1],
    label="zoro_acados_custom_update",
    linewidth=3,
)
plt.plot(
    data_zero_order_gpmpc["X"][:, 0],
    data_zero_order_gpmpc["X"][:, 1],
    label="zero_order_gpmpc",
    linewidth=1,
)
plt.legend()
```

```python
P_cup_arr = data_zoro_acados_custom_update["P"]
P_cup_arr
```

```python
N = data_zoro_acados_custom_update["X"].shape[0] - 1
P_cup = []
for i in range(N + 1):
    P_cup.append(
        np.array(
            [
                [P_cup_arr[3 * i], P_cup_arr[3 * i + 2]],
                [P_cup_arr[3 * i + 2], P_cup_arr[3 * i + 1]],
            ]
        )
    )

P_cup
```

```python
fig, ax = base_plot(lb_theta=0.0)

plot_data_zoro_acados = EllipsoidTubeData2D(
    center_data=data_zoro_acados["X"],
    ellipsoid_data=np.array(data_zoro_acados["P"]),
    # ellipsoid_data=None,
)
plot_data_zoro_acados_cupdate = EllipsoidTubeData2D(
    center_data=data_zoro_acados_custom_update["X"],
    ellipsoid_data=np.array(P_cup),
    # ellipsoid_data=None,
)
add_plot_trajectory(ax, plot_data_zoro_acados, color_fun=plt.cm.Greens, linewidth=3)
add_plot_trajectory(
    ax, plot_data_zoro_acados_cupdate, color_fun=plt.cm.Oranges, linewidth=1
)
```
